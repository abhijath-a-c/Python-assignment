### a function so that the columns of the output matrix are powers of the input vector.

```python
def vandermonde(v):
    vm=[]
    N=len(v)
    for i in v:
        R=[]
        for j in range(N):
            P=N-j-1
            R.append(int(i)**P)
        vm.append(R)
    return vm
```

```python
a=[1, 2, 3, 5]  
vandermonde(a)
```

### to find moving average in an array over a window

```python
def mvgavg(l,k):
    n=len(l)
    c=[sum(l[i:i+k]) for i in range(n-k+1)]
    ma=[round(i/k,2) for i in c]
    return ma
```

```python
a=[3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
window=3
```

```python
mvgavg(a,window)
```
