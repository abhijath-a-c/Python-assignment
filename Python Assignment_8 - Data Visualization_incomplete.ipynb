```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn import datasets
```

```python
ir=datasets.load_iris()
```

```python
ir.keys()
```

```python
print(ir['DESCR'])
```

```python
iris=pd.DataFrame(ir['data'],columns=ir['feature_names'])
```

```python
iris.head()
```

```python
iris.describe()
```

### Scaling data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(iris)
```

```python
scaled_data=scaler.transform(iris)
```

```python
scaled_data.shape
```

```python
type(scaled_data)
```

```python
plt.plot(scaled_data)
```

### Reducing Dimension

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
```

```python
pca.fit(scaled_data)
```

```python
pca_data=pca.transform(scaled_data)
```

```python
pca_data.shape
```

```python
type(pca_data)
```

### Plotting 

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.scatter(pca_data[:,0],pca_data[:,1],pca_data[:,2],c=ir['target'],cmap="plasma")
```
