### Python Program(with class concepts) to find the area of the triangle

```python
import math
class sides:
    def __init__(self):
        self.l=[]
    def length(self):
        for i in range(3):
            a=int(input("Enter length"))
            (self.l).append(a)
class area(sides):
    def area_c(sel):
        s=sum(sel.l)/2
        f=s
        for i in sel.l:
            f=f*(s-i)
        ar=math.sqrt(f)
        return ar
```

```python
ob=area()
ob.length()
print("Area is: ",ob.area_c())
```

### a function filter_long_words() 
##### takes a list of words and an integer n 
##### returns the list of words that are longer than n.

```python
def filter_long_words(l,n):
    up_l=[]
    for i in l:
        if len(i)>n:
            up_l.append(i)
    return up_l
```

#### Function that takes a list of words 
#### returns the list of intergers(len)

```python
def word_len(l):
    up_l=[]
    for i in l:
        up_l.append(len(i))
    return up_l
```

### function which takes a character and check vowel or not

```python
def vowel(c):
    l=['a','e','i','o','u']
    if c in l:
        return True
    else:
        return False
```
