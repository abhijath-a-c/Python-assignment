```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df
```

### Missing values in flight number

```python
df['FlightNumber']=[df['FlightNumber'][i-1]+10 if pd.isna(df['FlightNumber'][i]) else df['FlightNumber'][i] for i in range(len(df['FlightNumber']))]
```

```python
df
```

### From_to column into seperate columns

```python
ft=list(df['From_To'].str.split('_'))
```

```python
fromto=pd.DataFrame(ft,columns=['From','To'])
```

```python
fromto
```

### Standardize City name

```python
fromto_st=fromto.apply(lambda x:x.str.capitalize())
```

```python
fromto_st
```

### Delete from_to columns in dataframe

```python
df.pop('From_To')
```

```python
df
```

### Concatenate two dataframe

```python
df=pd.concat([fromto_st,df],axis=1)
```

```python
df
```

 ### Recent delays

```python
col=[]
for i in range(max(df['RecentDelays'].apply(len))):
    col.append("delay_"+str(i+1))
```

```python
delays=pd.DataFrame(df['RecentDelays'].to_list(),columns=col)
```
