```python
#Program to mimic reduce function
def myreduce(f,it):
    a=list(it)
    l=len(it)
    for i in range(1,l):
        a[i]=f(a[i-1],a[i])
    return a[l-1]
```

```python
#program to mimic filter function
def myfilter(f,it):
    a=list(it)
    b=[]
    l=len(it)
    for i in range(0,l):
        if f(a[i])==True:
            b.append(a[i])
            yield a[i]
```
