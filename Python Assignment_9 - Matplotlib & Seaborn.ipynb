### Plotting min and max of temperatures in a city

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
Max =np.array( [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
month=np.arange(0,12)
```

### Let  the periodic function be y = A sin(Bx+C)+D

```python
# Periodic fuction
def fitting_function(x,a,b,c,d):
    return (a*np.sin(b*(x)+c)+d)
```

```python
#fitting maximum temperature curve
p_max,pcon_max=curve_fit(fitting_function,month,Max)
```

```python
#fitting minimum temperature curve
p_min,pcon_min=curve_fit(fitting_function,month,Min)
```

```python
#generate dummy data for plotting curve
d=np.linspace(0,12.5,500)
```

```python
#plotting
plt.plot(month,Max,"ro")
plt.plot(d,fitting_function(d,*p_max),'r-')
plt.plot(month,Min,"bo")
plt.plot(d,fitting_function(d,*p_min),'b-')
plt.xlabel("Month")
plt.ylabel('Temperature ($^\circ$C)')
```

## Visualizing Titanic Data Set

```python
import pandas as pd
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
```

```python
titanic = pd.read_csv(url)
```

```python
titanic.head()
```

```python
fm=titanic['sex'].value_counts()
```

```python
fm.index
```

```python
plt.figure(figsize=(4,4))
plt.pie(fm.values,labels=fm.index.str.capitalize(),autopct='%0.2f%%')
plt.title('Male/Female proportion on Titanic',loc='center',size=14,font="monospace")
plt.legend(bbox_to_anchor=(1.3,1))
```

```python
m_fare=titanic[titanic.sex=='male']
f_fare=titanic[titanic.sex=='female']
```

```python
fig=plt.figure(figsize=(12,5))
plt.scatter(m_fare['fare'],m_fare['age'],c='r',label="Male")
plt.scatter(f_fare.fare,f_fare.age,c='b',label="Female")
plt.legend()
plt.xlabel("Age",font="Times New Roman")
plt.ylabel("Fare",font="Times New Roman")
```
